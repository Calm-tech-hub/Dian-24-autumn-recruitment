```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from matplotlib import pyplot as plt
import numpy as np
import logging
from tqdm import tqdm

# U-Net 模型 用于预测噪声，后续会利用MSE损失函数进行训练
class UNet(nn.Module):
    def __init__(self):
        super(UNet, self).__init__()
        self.encoder1 = self.double_conv(1, 64)#encoder1 是一个包含两个卷积层的模块，它的作用是从输入的图像中提取特征，将灰度的一个通道变为64个
        self.pool1 = nn.MaxPool2d(2)#最大池化层通常与编码器层交替使用，以逐步提取图像的特征并减少特征图的尺寸
        self.encoder2 = self.double_conv(64, 128)
        self.pool2 = nn.MaxPool2d(2)
        self.middle = self.double_conv(128, 256)
        self.upconv2 = self.up_conv(256, 128)#转置卷积的操作会将输入特征图的高和宽扩大一倍。这种特征融合有助于模型保留低层次特征，同时结合高层次特征，进一步提高重建的准确性和细节
        self.decoder2 = self.double_conv(256, 128)
        self.upconv1 = self.up_conv(128, 64)#这种结构有助于在上采样过程中保留低层次的细节信息。
        self.decoder1 = self.double_conv(128, 64)
        self.out_conv = nn.Conv2d(64, 1, kernel_size=1)

    def double_conv(self, in_channels, out_channels):#两个卷积加上两个激活函数
        return nn.Sequential(
            nn.Conv2d(in_channels, out_channels, kernel_size=3, padding=1),
            nn.ReLU(inplace=True),#增加非线性特征，并提高模型的表达能力。
            nn.Conv2d(out_channels, out_channels, kernel_size=3, padding=1),
            nn.ReLU(inplace=True)
        )

    def up_conv(self, in_channels, out_channels):#将特征图的空间维度增大，以便在解码阶段逐步恢复到原始图像的尺寸
        return nn.ConvTranspose2d(in_channels, out_channels, kernel_size=2, stride=2)

    def forward(self, x):
        e1 = self.encoder1(x)
        e2 = self.encoder2(self.pool1(e1))
        m = self.middle(self.pool2(e2))
        d2 = self.upconv2(m)
        d2 = torch.cat([d2, e2], dim=1)#跳跃连接
        d2 = self.decoder2(d2)
        d1 = self.upconv1(d2)
        d1 = torch.cat([d1, e1], dim=1)#跳跃连接
        d1 = self.decoder1(d1)
        out = self.out_conv(d1)
        return out



# RNN 模型定义（用于分类任务），这里就用简化版的RNN
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size, output_size, num_layers=1):
        super(RNN, self).__init__()
        self.rnn = nn.RNN(input_size, hidden_size, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out, _ = self.rnn(x)
        out = self.fc(out[:, -1, :])  # 输出最后一个时间步的结果
        return out

# 评估RNN模型
def evaluate_rnn(test_loader, model, device):
    model.eval()#将模型变为评估模式
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in test_loader:
            images = images.to(device)
            labels = labels.to(device)

            # 首先调整图片的维度
            images_flat = images.view(images.size(0), -1) 
            images_flat = images_flat.unsqueeze(1)  
            outputs = model(images_flat)
            _, predicted = torch.max(outputs.data, 1)#取预测概率最大的
            total += labels.size(0)#取该批次的个数
            correct += (predicted == labels).sum().item()

    accuracy = 100 * correct / total#计算准确率
    logging.info(f'RNN Test Accuracy: {accuracy:.2f}%')
    model.train() 

# 获取数据
def get_data(args):
    transform = transforms.Compose([
        transforms.Resize((args.image_size, args.image_size)),
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,)),
    ])
    dataset = datasets.FashionMNIST(root='./data', train=True, download=True, transform=transform)
    dataloader = DataLoader(dataset, batch_size=args.batch_size, shuffle=True)
    return dataloader


# 获取测试数据
def get_test_data(args):
    transform = transforms.Compose([
        transforms.Resize((args.image_size, args.image_size)),
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,)),
    ])
    test_dataset = datasets.FashionMNIST(root='./data', train=False, download=True, transform=transform)
    test_dataloader = DataLoader(test_dataset, batch_size=args.batch_size, shuffle=False)
    return test_dataloader

# 扩散模型类
class Diffusion:
    def __init__(self, noise_steps=500, beta_start=1e-4, beta_end=0.02, img_size=28, device="cuda"):
        self.noise_steps = noise_steps#将原文中的1000次减为500次
        self.beta_start = beta_start
        self.beta_end = beta_end
        self.img_size = img_size
        self.device = device

        self.beta = self.prepare_noise_schedule().to(device)
        self.alpha = 1. - self.beta
        self.alpha_hat = torch.cumprod(self.alpha, dim=0).to(device)  # 计算alpha的累乘

    def prepare_noise_schedule(self):
        return torch.linspace(self.beta_start, self.beta_end, self.noise_steps)#生成beta序列，从1e-4到0.02，与原论文中数据保持一致

    def noise_images(self, x, t):
        sqrt_alpha_hat = torch.sqrt(self.alpha_hat[t])[:, None, None, None]#计算根号alpha
        sqrt_one_minus_alpha_hat = torch.sqrt(1 - self.alpha_hat[t])[:, None, None, None]#根号下1-alpha
        noise = torch.randn_like(x)
        return sqrt_alpha_hat * x + sqrt_one_minus_alpha_hat * noise, noise#返回下一个时间步的x_t,以及生成的随机高斯分布的噪声

    def sample_timesteps(self, n):
        return torch.randint(low=1, high=self.noise_steps, size=(n,), device=self.device)  # 生成时间步数量

    def sample(self, model, n):#进行采样，其中model是unet，即在这个采样过程，也就是反向传播过程中，用unet函数进行预测噪声
        logging.info(f"Sampling {n} new images....")
        model.eval()
        with torch.no_grad():
            x = torch.randn((n, 1, self.img_size, self.img_size), device=self.device)
            for i in tqdm(reversed(range(1, self.noise_steps)), position=0):#tqdm生成进度条
                t = torch.tensor([i] * x.size(0), device=self.device).long()
                predicted_noise = model(x)  # 用unet预测噪声
                alpha = self.alpha[t][:, None, None, None]
                alpha_hat = self.alpha_hat[t][:, None, None, None]#alpha的累乘
                beta = self.beta[t][:, None, None, None]
                if i > 1:
                    noise = torch.randn_like(x)
                else:
                    noise = torch.zeros_like(x)
                x = 1 / torch.sqrt(alpha) * (x - ((1 - alpha) / (torch.sqrt(1 - alpha_hat))) * predicted_noise) + torch.sqrt(beta) * noise#反向传播的公式，计算前一个时间步的x_t
        model.train()#将模型设为训练模式
        x = (x.clamp(-1, 1) + 1) / 2#对张量 x 的值进行归一化处理，使其从原本的范围转换到 [0, 1] 之间
        x = (x * 255).cpu().numpy().astype(np.uint8)#这段代码的作用是将张量 x 转换为 NumPy 数组，并将其值从 [0, 1] 的范围转换为 [0, 255] 的范围，适合用作图像数据。具体步骤如下：
        return x


# 训练函数
def train_rnn(args):
    
    device = torch.device(args.device if torch.cuda.is_available() else "cpu")
    dataloader = get_data(args)
    test_dataloader = get_test_data(args)

    # 初始化模型
    unet_model = UNet().to(device)
    rnn_model = RNN(input_size=args.image_size * args.image_size, hidden_size=128, output_size=10).to(device)
    optimizer = optim.Adam(list(unet_model.parameters()) + list(rnn_model.parameters()), lr=args.lr)#对unet以及rnn模型的优化器
    mse = nn.MSELoss()#这是unet训练的损失函数，均方误差函数
    CE=nn.CrossEntropyLoss()#这是rnn训练的损失函数，交叉熵函数
    diffusion = Diffusion(img_size=args.image_size, device=device)

    for epoch in range(args.epochs):
        logging.info(f"Starting epoch {epoch}:")
        pbar = tqdm(dataloader)
        for i, (images, labels) in enumerate(pbar):
            images = images.to(device)
            labels = labels.to(device)

            # 优化Unet模型
            t = diffusion.sample_timesteps(images.size(0)).to(device)
            x_t, noise = diffusion.noise_images(images, t)
            predicted_noise = unet_model(x_t)
            loss = mse(noise, predicted_noise) #unet网络的损失函数

            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            # 优化RNN模型
            #先对输入图像的维度进行处理
            images_flat = images.view(images.size(0), -1)  
            images_flat = images_flat.unsqueeze(1)
            rnn_output = rnn_model(images_flat)
            rnn_loss = CE(rnn_output, labels)  # RNN的损失函数

            optimizer.zero_grad()
            rnn_loss.backward()
            optimizer.step()

            pbar.set_postfix(MSE=loss.item(), RNN_Loss=rnn_loss.item())

        if epoch % 4 == 0: #进行可视化以及性能评估
            generate_and_display_images(args, unet_model)
            evaluate_rnn(test_dataloader, rnn_model, device)


# 生成和显示图片
def generate_and_display_images(args, model):
    device = torch.device(args.device if torch.cuda.is_available() else "cpu")
    diffusion = Diffusion(img_size=args.image_size, device=device)
    
    # Generate images
    sampled_images = diffusion.sample(model, n=8)
    
    plt.figure(figsize=(10, 5))
    for i in range(8):
        img = sampled_images[i]
        img = img.squeeze(0)  #处理灰度图像，只有一个通道
        # 将像素值从范围 [0, 1] 转换为 [0, 255]
        img = (img * 255).astype(np.uint8)
        plt.subplot(2, 4, i + 1)
        plt.imshow(img, cmap='gray')
        plt.axis('off')
    
    plt.show()



# 启动函数
def launch():
    import argparse

    # 参数设置
    class ARGS:
        def __init__(self):
            self.run_name = "DDPM_RNN"
            self.epochs = 101
            self.batch_size = 64
            self.image_size = 28
            self.dataset_path = './data'
            self.device = "cuda" if torch.cuda.is_available() else "cpu"
            self.lr = 1.5e-4

    args = ARGS()
    train_rnn(args)

if __name__ == '__main__':
    launch()
```
